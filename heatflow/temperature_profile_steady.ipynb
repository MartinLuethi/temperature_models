```python
%matplotlib inline
import numpy as np
import pylab as plt

hmax = 350.
G    = 0.02   # temperature gradient K/m
T_s  = -10
k    = 2   # conductivity

G = -0.07/k

h = np.linspace(0,hmax,50)
T = G*(-h) + T_s
Q = k*G



print('Heat flux = %.1f mW/m2' %(Q*1000))
plt.figure(figsize=(8,8))
plt.plot(T, -h, 'r')
plt.xlabel('Temperature ($^\circ$C)')
plt.ylabel('Depth (m)')

plt.show()
```
